read and visualize .vtp mesh files

```python
import vtk

def read_vtp_file(file_path):
    # 创建一个vtkXMLPolyDataReader对象
    reader = vtk.vtkXMLPolyDataReader()
    # 设置输入文件名
    reader.SetFileName(file_path)
    # 读取文件
    reader.Update()

    # 获取读取的数据
    polydata = reader.GetOutput()
    return polydata

def visualize_surface(polydata):
    # 创建一个vtkPolyDataMapper对象，并设置输入数据
    mapper = vtk.vtkPolyDataMapper()
    mapper.SetInputData(polydata)

    # 创建一个vtkActor对象，并将Mapper设置为该Actor的Mapper
    actor = vtk.vtkActor()
    actor.SetMapper(mapper)

    # 创建一个vtkRenderer对象，并将Actor添加到该Renderer
    renderer = vtk.vtkRenderer()
    renderer.AddActor(actor)

    # 创建一个vtkRenderWindow对象，并将Renderer设置为该RenderWindow的Renderer
    render_window = vtk.vtkRenderWindow()
    render_window.AddRenderer(renderer)

    # 创建一个vtkRenderWindowInteractor对象，用于处理交互操作
    interactor = vtk.vtkRenderWindowInteractor()
    interactor.SetRenderWindow(render_window)

    # 启动可视化
    render_window.Render()
    interactor.Start()


if __name__ == "__main__":
    # 替换为您的.vtp文件路径
    vtp_file_path = "./data/original_data/remeshed/area-001/cut2/C0001_cut2.vtp"
    polydata = read_vtp_file(vtp_file_path)

    # 可以在这里对读取的数据进行处理或显示
    visualize_surface(polydata)
    print(polydata)
```
